```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "notebook"

# Select file
filename = "/work/clas12/users/touchte/amon/python-utils/rcdb/output/results.csv" 

# ignore comment, line starting with #
df = pd.read_csv(filename, skipinitialspace=True, comment="#")

# show the 20 first lines
df[0:20]
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(10, 5), dpi=150)

plt.scatter(x=df["run"], y=df["torus"])
plt.xlabel("run")
plt.ylabel("Torus scale")
plt.title("Torus scale over the run period")

ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))


plt.tight_layout()
# plt.savefig("./output/torus_sacle.pdf", bbox_inches="tight")
# plt.savefig("./output/torus_scale.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Interactive figure
fig1 = px.scatter(x=df["run"], y=df["torus"])

fig1.update_layout(
    title={
        "text": "Torus scale over the run period",
        "x": 0.5,
        "xanchor": "center",
    },
    xaxis_title="Run number",
    yaxis_title="Torus scale",
    width=1000,
    height=500
)

fig1.show()
```
