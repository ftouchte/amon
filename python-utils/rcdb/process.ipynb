```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "notebook"

# Select file
filename = "/work/clas12/users/touchte/amon/python-utils/rcdb/output/results.csv" 

# ignore comment, line starting with #
df = pd.read_csv(filename, skipinitialspace=True, comment="#")

# show the 20 first lines
df[0:20]
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(10, 5), dpi=150)

plt.scatter(x=df["run"], y=df["torus"])
plt.xlabel("run")
plt.ylabel("Torus scale")
plt.title("Torus scale over the run period")

ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))


plt.tight_layout()
# plt.savefig("./output/torus_sacle.pdf", bbox_inches="tight")
# plt.savefig("./output/torus_scale.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Interactive figure

df["dvz1"] = abs(10*df["mean"] + df["alignment"])

fig1 = px.scatter(x=-df["alignment"], y=df["dvz1"])  # optionnel

fig1.update_layout(
    title={
        "text": "abs(mean-alignment) versus alignment",
        "x": 0.5,
        "xanchor": "center",
    },
    xaxis_title="alignment (mm)",
    yaxis_title="abs(mean - alignment) (mm)",
    width=600,
    height=500
)

fig1.show()
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(6, 5), dpi=150)

plt.scatter(-df["alignment"], df["width"])
plt.xlabel("vz shift (mm)")
plt.ylabel("width (cm)")
plt.title(r"Width of $(vz^{CLAS,e} - vz^{AHDC})$ versus shift")

# only show integer in the x axis
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout()
plt.savefig("./output/alignment_vs_width.pdf", bbox_inches="tight")
plt.savefig("./output/alignment_vs_width.png", dpi=300, bbox_inches="tight")
plt.show()
```
