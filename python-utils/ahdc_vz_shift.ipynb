# Routine to select the optimal vz shift for a given data set

1. Run the `new_ahdc_position_scan()` the **AhdcAlignmentAnalyser**
1. A file results.csv will be created, set the path of this file in the cell below


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "notebook"

# Select file
# 10.6 GeV
filename = "/volatile/clas12/touchte/position-scan-10gev/results.csv"
#2.24 GeV
#filename = "/volatile/clas12/touchte/position-scan-2gev/results.csv" 

# ignore comment, line starting with #
df = pd.read_csv(filename, skipinitialspace=True, comment="#")

# show the 20 first lines
df[0:20]
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(6, 5), dpi=150)

plt.scatter(-df["alignment"], abs(10*df["mean"] + df["alignment"]))
plt.xlabel("vz shift (mm)")
plt.ylabel(r"$|\langle  vz^{CLAS,e} - vz^{AHDC} \rangle - vz^{shift}|$  (mm)")
plt.title("Optimal vz shift")
#plt.colorbar(label="layer")

# Force les labels à être des entiers (le pas est choisi automatiquement)
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))


plt.tight_layout()
plt.savefig("./output/optimal_vz_shift.pdf", bbox_inches="tight")
plt.savefig("./output/optimal_vz_shift.png", dpi=300, bbox_inches="tight")
plt.show()
plt.show()
```

```python
# Interactive figure

df["dvz1"] = abs(10*df["mean"] + df["alignment"])

fig1 = px.scatter(x=-df["alignment"], y=df["dvz1"])  # optionnel

fig1.update_layout(
    title={
        "text": "abs(mean-alignment) versus alignment",
        "x": 0.5,
        "xanchor": "center",
    },
    xaxis_title="alignment (mm)",
    yaxis_title="abs(mean - alignment) (mm)",
    width=600,
    height=500
)

fig1.show()
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(6, 5), dpi=150)

plt.scatter(-df["alignment"], df["width"])
plt.xlabel("vz shift (mm)")
plt.ylabel("width (cm)")
plt.title(r"Width of $(vz^{CLAS,e} - vz^{AHDC})$ versus shift")

# only show integer in the x axis
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout()
plt.savefig("./output/alignment_vs_width.pdf", bbox_inches="tight")
plt.savefig("./output/alignment_vs_width.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Interactive figure
fig = px.scatter(x=-df["alignment"], y=df["width"])  # optionnel

fig.update_layout(
    title={
        "text": "Distribution width versus vz shift",
        "x": 0.5,
        "xanchor": "center",
    },
    xaxis_title="vz shift (mm)",
    yaxis_title="width (cm)",
    width=600,
    height=500
)

fig.show()
```
