```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "notebook"

filename = "./files/wire_angles.csv"
df = pd.read_csv(filename, skipinitialspace=True, comment="#")

df["num"] = df.index

df[0:20]

#df.describe()
```

```python
fig = px.scatter(df, x ="num", y="angle",
                 color="layer",
                 color_continuous_scale="Viridis")  # optionnel

fig.update_layout(
    title={
        "text": "Angle versus wire angle",
        "x": 0.5,
        "xanchor": "center"

    },
    xaxis_title="wire num",
    yaxis_title="angle (deg)"
)

fig.show()
```
